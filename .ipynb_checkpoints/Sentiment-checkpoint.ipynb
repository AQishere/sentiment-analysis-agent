```python
#Import Libraries, Load Dataset and Clean Dataset
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
import nltk
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report, accuracy_score

# Define the data types for each column
data_types = {
    'timestamp': int,
    'date_time': str,  # Corrected column name
    'username': str,
    'handle': str,
    'url': str,
    'text': str
}

# Define a function to convert the date_time column to datetime objects
def parse_date_time(date_str):
    return pd.to_datetime(date_str, format='%a %b %d %H:%M:%S PDT %Y')

# Load the dataset with the correct encoding and data types
df = pd.read_csv('training.1600000.processed.noemoticon.csv', encoding='ISO-8859-1', dtype=data_types)

# Apply the date_time parsing function to the 'date_time' column
df['date_time'] = df['Mon Apr 06 22:19:45 PDT 2009'].apply(parse_date_time)

# Display the first few rows of the dataframe
print(df.head())
```

```python
#Explore Data (EDA)
```

```python
# Check the shape of the DataFrame
print(f"Shape of the DataFrame: {df.shape}")

# Get a summary of the dataset
print(df.info())

# Check for missing values
print(df.isnull().sum())

# Count the number of positive and negative reviews (assuming 'NO_QUERY' indicates sentiment)
print(df['NO_QUERY'].value_counts())
```

```python
#Clean Text
```

```python
import re
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer

# Load NLTK data
nltk.download('stopwords')
nltk.download('punkt')

# Initialize stop words and stemmer
stop_words = set(stopwords.words('english'))
stemmer = PorterStemmer()

def clean_text(text):
    # Remove HTML tags (if any)
    text = re.sub(r'<.*?>', ' ', text)
    
    # Remove non-alphabetic characters
    text = re.sub(r'[^a-zA-Z]', ' ', text)
    
    # Convert to lowercase
    text = text.lower()
    
    # Split into words and stem each word
    words = [stemmer.stem(word) for word in text.split() if word not in stop_words]
    
    # Join words back into a string
    return ' '.join(words)

# Apply the clean_text function to the reviews (assuming 'NO_QUERY' contains the text)
df['clean'] = df['NO_QUERY'].apply(clean_text)
```

```python
#TF-IDF Vectorization
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split

# Initialize the TF-IDF vectorizer
tfidf = TfidfVectorizer(max_features=10000, ngram_range=(1, 2), min_df=2)

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df['clean'], df['NO_QUERY'].map(lambda x: 1 if 'positive' in x.lower() else 0), test_size=0.2, random_state=42)

# Fit and transform the training data
X_train_vec = tfidf.fit_transform(X_train)

# Transform the test data using the same vectorizer
X_test_vec = tfidf.transform(X_test)
```

```python
#Train Models
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report, accuracy_score

# Encode labels (assuming 'NO_QUERY' contains the sentiment)
y_train = df['NO_QUERY'].map(lambda x: 1 if 'positive' in x.lower() else 0)

# Train Logistic Regression model
lr_model = LogisticRegression(C=1.0, max_iter=300)
lr_model.fit(X_train_vec, y_train)  # Ensure X_train_vec and y_train are aligned

# Predict and evaluate
lr_preds = lr_model.predict(X_test_vec)
print(f"Logistic Regression Accuracy: {accuracy_score(y_test, lr_preds):.2%}")
print(classification_report(y_test, lr_preds))

# Train Naive Bayes model
nb_model = MultinomialNB(alpha=0.1)
nb_model.fit(X_train_vec, y_train)  # Ensure X_train_vec and y_train are aligned

# Predict and evaluate
nb_preds = nb_model.predict(X_test_vec)
print(f"Naive Bayes Accuracy: {accuracy_score(y_test, nb_preds):.2%}")
print(classification_report(y_test, nb_preds))
```
